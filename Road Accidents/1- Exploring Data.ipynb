```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.dates as mdates
import matplotlib.colors as mcolors
import difflib
```

# Data Exploration 2023

## Read CSVs

```python
caract_2023 = pd.read_csv("caract-2023.csv",encoding="utf-8",sep=";")
lieux_2023 = pd.read_csv("lieux-2023.csv",encoding="utf-8",sep=";")
usagers_2023 = pd.read_csv("usagers-2023.csv",encoding="utf-8",sep=";")
vehicules_2023 = pd.read_csv("vehicules-2023.csv",encoding="utf-8",sep=";")
```

### Characteristics

```python
display(caract_2023.head(20))
print()
print(caract_2023.info())
print()
print(caract_2023.describe())
print()
print(caract_2023.nunique())
print()
print(caract_2023['lum'].value_counts())
print()
print(caract_2023['agg'].value_counts())
print()
print(caract_2023['int'].value_counts())
print()
print(caract_2023['atm'].value_counts())
print()
print(caract_2023['col'].value_counts())
```

Lighting conditions at accident time (lum): most of the accidents occurred during daylight (1)
Localisation (Agg): Most accidents occurred in urban areas (2)
Type of Intersection (Int): Most accidents occurred not at intersections (1)
Meteorological factors (atm): Most accidents occurred in normal weather (1)

### Places

```python
display(lieux_2023.head(15))
print()
print(lieux_2023.info())
print()
print(lieux_2023.describe())
print()
print(lieux_2023.nunique())
print()
print(lieux_2023['catr'].value_counts())
print()
print(lieux_2023['voie'].value_counts())
print()
print(lieux_2023['circ'].value_counts())
print()
print(lieux_2023['nbv'].value_counts())
print()
print(lieux_2023['surf'].value_counts())
print()
print(lieux_2023['situ'].value_counts())
print()
print(lieux_2023['vma'].value_counts())
```

### Users

```python
display(usagers_2023.head(20))
print(usagers_2023.info())

# grav = Severity (The Target Variable)
# 1=Uninjured, 2=Killed, 3=Hospitalized, 4=Slightly Injured
print("Severity Distribution (Target):\n", usagers_2023['grav'].value_counts())

# catu = Category of User (1=Driver, 2=Passenger, 3=Pedestrian)
print("User Category:\n", usagers_2023['catu'].value_counts())

# secu1, secu2, secu3 = Safety Equipment (Seatbelt, Helmet)
# This often requires cleaning (values like 11, 21 often mean 'Belt Used', 'Helmet Used')
print("Safety Equipment:\n", usagers_2023['secu1'].value_counts())
```

### Vehicles

```python
display(vehicules_2023.head())
print(vehicules_2023.info())
```

```python
# catv = Category of Vehicle
# Common codes: 7=Car, 10=Van, 33=Motorcycle (heavy), 1=Bicycle
print("Vehicle Categories:\n", vehicules_2023['catv'].value_counts())

# obs = Obstacle hit (fixed)
# 0=None, 1=Tree, 2=Pole, etc. useful for severity prediction.
print("Fixed Obstacles Hit:\n", vehicules_2023['obs'].value_counts())
```
